```python
import mysql.connector
import pandas as pd
import csv
```

Supongamos que tenemos 7 días a la semana y para el restaurante hay jornada de mediodía y jornada de noche. De lunes a viernes en la jornada de mediodía se necesita 1 cocinero 1 barrero y 1 salero, y en la jornada de noche 1 cocinero 2 barreros y 2 saleros. En la jornada de mediodía los fines de semana se necesitan 1 cocinero 2 barreros y 2 saleros y en la jornada de noche de los fines de semana se necesitan 2 cocineros 2 barreros y 3 saleros. ¿Como se distribuyen las horas de trabajo?

```python
# # Import all classes of PuLP module
from pulp import *

listrabajadores = [0,'Cocinero1','Cocinero2','Barrero1','Barrero2','Barrero3','Salero1','Salero2','Salero3','Salero4']
workers=[1,2,3,4,5,6,7,8,9]
dias_turnos = ['Lunes_M', 'Lunes_N', 'Martes_M', 'Martes_N', 'Miércoles_M', 'Miércoles_N','Jueves_M', 'Jueves_N', 'Viernes_M', 'Viernes_N', 'Sábado_M', 'Sábado_N','Domingo_M', 'Domingo_N']

secciones = ['Cocina', 'Barra', 'Sala']

jobs_labels = [dt + "_" + sec for sec in secciones for dt in dias_turnos]

jobs=[i+1 for i in range(len(jobs_labels))]

# Cost Matrix
costs=[[1 for i in jobs] for j in workers]

prob = LpProblem("Assignment Problem", LpMinimize) 
```

```python
# The cost data is made into a dictionary
costs= makeDict([workers, jobs], costs, 0)

# Creates a list of tuples containing all the possible assignments
assign = [(w, j) for w in workers for j in jobs]

# A dictionary called 'Vars' is created to contain the referenced variables
vars = LpVariable.dicts("Assign", (workers, jobs), 0, None, LpBinary)
```

```python
# The objective function is added to 'prob' first
prob += (
    lpSum([vars[w][j] * costs[w][j] for (w, j) in assign]),
    "Sum_of_Assignment_Costs",
)
```

```python
# # Requerimientos mínimos por turno
diarioM = [1,3,5,7,9]
diarioN = [2,4,6,8,10]
findesM = [11,13]
findesN = [12,14]
#Entre diario
#Mediodía
#Mínimo 1 cocinero
for dia in diarioM:
    prob+= lpSum([vars[1][dia],vars[2][dia]]) >= 1
#Mínimo 1 barrero
for dia in diarioM:
    prob+= lpSum([vars[3][dia],vars[4][dia],vars[5][dia]]) >= 1
#Mínimo 1 salero
for dia in diarioM:
    prob+= lpSum([vars[6][dia],vars[7][dia],vars[8][dia],vars[9][dia]]) >= 1
#Noche
#Mínimo 1 cocinero
for dia in diarioN:
    prob+= lpSum([vars[1][dia],vars[2][dia]]) >= 1
#Mínimo 2 barreros
for dia in diarioN:
    prob+= lpSum([vars[3][dia],vars[4][dia],vars[5][dia]]) >= 2
#Mínimo 2 saleros
for dia in diarioN:
    prob+= lpSum([vars[6][dia],vars[7][dia],vars[8][dia],vars[9][dia]]) >= 2

#Fines de semana
#Mediodía
#Mínimo 1 cocinero
for dia in findesM:
    prob+= lpSum([vars[1][dia],vars[2][dia]]) >= 1
#Mínimo 2 barreros
for dia in findesM:
    prob+= lpSum([vars[3][dia],vars[4][dia],vars[5][dia]]) >= 2
#Mínimo 2 saleros
for dia in findesM:
    prob+= lpSum([vars[6][dia],vars[7][dia],vars[8][dia],vars[9][dia]]) >= 2
#Noche
#Mínimo 2 cocineros
for dia in findesN:
    prob+= lpSum([vars[1][dia],vars[2][dia]]) >= 2
#Mínimo 2 barreros
for dia in findesN:
    prob+= lpSum([vars[3][dia],vars[4][dia],vars[5][dia]]) >= 2
#Mínimo 3 saleros
for dia in findesN:
    prob+= lpSum([vars[6][dia],vars[7][dia],vars[8][dia],vars[9][dia]]) >= 3
```

```python
# The problem is solved using PuLP's choice of Solver
prob.solve()

varValues = []
# Print the variables optimized value
for v in prob.variables():
    varValues.append(v.varValue)
    
# The optimised objective function value is printed to the screen
print("Value of Objective Function = ", value(prob.objective))
```

```python
separa=[str(prob.variables()[i]).split('_') for i in range(len(prob.variables()))]
for i in range(len(prob.variables())):
    separa[i]=['Assign',int(separa[i][1]),int(separa[i][2]),int(varValues[i])]
jobsepara=[]
for i in range(len(jobs_labels)):
    jobsepara.append(jobs_labels[i].split('_'))
```
